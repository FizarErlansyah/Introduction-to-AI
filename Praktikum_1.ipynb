```python
#1. Import Library AI
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets
```

```python
#2. Load Dataset
digits = datasets.load_digits()
print("Jumlah data:", len(digits.data))
print("Jumlah fitur per data:", len(digits.data[0]))
```

```python
#3. Tampilkan Gambar
plt.imshow(digits.images[0], cmap='gray')
plt.title("Contoh Gambar Angka")
plt.show()
```

```python
#4. Latih Model Sederhana (Tanpa Teori)
from sklearn.linear_model import LogisticRegression
X = digits.data
y = digits.target
model = LogisticRegression(max_iter=1000)
model.fit(X, y)
print("Model berhasil dilatih!")
```

```python
#5. Minta Model Menebak
prediksi = model.predict(X[:10])
print("Prediksi:", prediksi)
print("Jawaban asli:", y[:10])
```
