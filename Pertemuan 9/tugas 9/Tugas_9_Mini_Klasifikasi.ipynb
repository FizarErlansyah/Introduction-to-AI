# Tugas 9: Mini Klasifikasi Berdasarkan Dataset Kelompok

## 2. Identifikasi Komponen Klasifikasi
- **Objek:** Data tanggapan responden dalam dataset Survei Penggunaan AI.
- **Label / Kelas:** Tingkat ketergantungan terhadap AI (merujuk pada kolom `4. Jika dinilai, masuk ke dalam klasifikasi manakah tingkat ketergantungan Anda terhadap AI?`).
- **Fitur:** Tujuan utama dalam menggunakan AI (kita fokus ke satu kolom berupa teks bebas `2. Apa tujuan utama Anda menggunakan AI saat ini?`).

## 3. Preprocessing Sederhana
Preprocessing yang dilakukan pada data teks agar siap diproses oleh model. Langkah yang diterapkan:
1. **Case Folding (Lowercase):** Mengubah seluruh huruf menjadi huruf kecil.
2. **Menghapus Tanda Baca:** Menghilangkan simbol atau tanda baca agar tidak dianggap sebagai fitur unik oleh perhitungan probabilitas (misalnya diubah menjadi regex tanpa simbol).

**Contoh Sebelum & Sesudah:**
- **Sebelum:** `"Mencari Ide / Brainstorming, Menyusun Draf Tulisan (Laporan, Artikel, Email)"`
- **Sesudah:** `"mencari ide  brainstorming menyusun draf tulisan laporan artikel email"`

### 1. Import Library

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
```

### 2. Load Dataset

```python
from IPython.display import display, HTML

# 2. Load Dataset
data = pd.read_csv("Survei Penggunaan AI.csv")

nama_kolom_fitur = "2. Apa tujuan utama Anda menggunakan AI saat ini? (Bisa pilih lebih dari satu)"
nama_kolom_label = '4. Jika dinilai, masuk ke dalam klasifikasi manakah tingkat ketergantungan Anda terhadap AI? (Pilih satu yang paling menggambarkan Anda) '

# Drop data kosong
data = data.dropna(subset=[nama_kolom_fitur, nama_kolom_label])

# Mengambil tepat 20 data sesuai dengan instruksi ("minimal 20 data saja")
data = data.head(20)

X = data[nama_kolom_fitur].copy()
y = data[nama_kolom_label].copy()

display(HTML(f"<b>Total Data Keseluruhan:</b> {len(data)} baris"))
display(HTML("<br><b>--- Menampilkan Seluruh (20) Data ---</b>"))

# Merapikan tabel dengan teks rata kiri (Left Aligned)
styled_data = data[[nama_kolom_fitur, nama_kolom_label]].style.set_properties(**{'text-align': 'left'}).set_table_styles([dict(selector='th', props=[('text-align', 'left')])])

# Menampilkan 20 data agar tampil semuanya 
with pd.option_context('display.max_rows', None, 'display.max_colwidth', None):
    display(styled_data)
```

### 3. Preprocessing Sederhana

```python
# 3. Preprocessing Sederhana
from IPython.display import display, HTML

# Salin bentuk asli (Sebelum)
X_sebelum = X.copy()

# Tahap 1: Case Folding (Mengubah semua huruf menjadi huruf kecil - Lowercase)
X_lowercase = X.astype(str).str.lower()

# Tahap 2: Menghapus tanda baca menggunakan regex (Hanya menyisakan alphabet, angka, serta spasi)
X_bersih = X_lowercase.str.replace(r'[^a-zA-Z0-9\s]', '', regex=True)

# Simpan hasil akhir kembali ke variabel X agar bisa dipakai di langkah selanjutnya
X = X_bersih.copy()

# Buat dataframe berisi proses step-by-step untuk 10 data pertama agar terlihat perbedaannya dengan jelas
df_banding = pd.DataFrame({
    "Teks Asli (Mentah)": X_sebelum.head(10),
    "Tahap 1 (Lowercase)": X_lowercase.head(10),
    "Tahap 2 (Tanpa Tanda Baca / Final)": X_bersih.head(10)
})

# Merapikan tabel hasil preprocessing dengan format rata kiri (Left Aligned)
styled_banding = df_banding.style.set_properties(**{'text-align': 'left'}).set_table_styles([dict(selector='th', props=[('text-align', 'left')])])

display(HTML("<b>======= DETAIL HASIL PREPROCESSING (10 Data Pertama) =======</b>"))
# Menampilkan tabel agar seluruh teks tidak terpotong
with pd.option_context('display.max_rows', None, 'display.max_colwidth', None):
    display(styled_banding)
```

### 4. Ubah Teks ke Angka (Vektorisasi)

```python
# 4. Ubah Teks ke Angka (Vektorisasi)
from IPython.display import display, HTML

vectorizer = CountVectorizer()
X_vector = vectorizer.fit_transform(X)

df_vektor_info = pd.DataFrame({
    "Dimensi Matriks": ["Jumlah Baris (Sampel)", "Jumlah Kolom (Kosakata Unik)"],
    "Nilai": [X_vector.shape[0], X_vector.shape[1]]
})

display(HTML("<b>Dimensi data teks setelah diubah ke angka (CountVectorizer):</b>"))
display(df_vektor_info)
```

### 5. Split Data (80% Train, 20% Test)

```python
# 5. Split Data (80% Train, 20% Test)
from IPython.display import display, HTML

X_train, X_test, y_train, y_test = train_test_split(X_vector, y, test_size=0.2, random_state=42)

df_split = pd.DataFrame({
    "Jenis Dataset": ["Latih (Training)", "Uji (Testing)"],
    "Proporsi (%)": [80, 20],
    "Total Baris": [X_train.shape[0], X_test.shape[0]]
})

display(HTML("<b>Informasi Pembagian (Split) Data:</b>"))
display(df_split)
```

### 6. Model Training & Prediksi Menggunakan Naive Bayes

```python
# 6. Model Training & Prediksi Menggunakan Naive Bayes
from IPython.display import display, HTML

model = MultinomialNB()
# Train model dengan fit
model.fit(X_train, y_train)

# Prediksi model
hasil = model.predict(X_test)

# Buat dataframe hasil biar rapi berupa tabel
arr_status = []
for i in range(len(hasil)):
    if y_test.values[i] == hasil[i]:
        arr_status.append("Sesuai ✅")
    else:
        arr_status.append("Beda ❌")

df_hasil_prediksi = pd.DataFrame({
    "Data Ke-": range(1, len(hasil) + 1),
    "Label Aktual Target (y_test)": y_test.values,
    "Hasil Prediksi AI": hasil,
    "Status": arr_status
})

# Merapikan tabel dengan teks rata kiri dan menghilangkan index bawaan pandas (angka 0, 1, 2...)
styled_prediksi = df_hasil_prediksi.style.hide(axis="index").set_properties(**{'text-align': 'left'}).set_table_styles([dict(selector='th', props=[('text-align', 'left')])])

display(HTML("<b>========= TABEL HASIL PREDIKSI =========</b>"))

# Menampilkan tabel agar seluruh teks tidak terpotong
with pd.option_context('display.max_rows', None, 'display.max_colwidth', None):
    display(styled_prediksi)
```
