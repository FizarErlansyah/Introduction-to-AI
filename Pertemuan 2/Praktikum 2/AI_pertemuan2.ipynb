Data set IPK, kehadiran dan status Lulus/Tidak Lulus

```python
import pandas as pd
data = pd.DataFrame({
'IPK': [3.5, 2.1, 3.8, 2.5, 3.2, 2.0, 3.6, 2.7, 2.0],
'Kehadiran': [90, 60, 95, 70, 85, 50, 92, 75, 75],
'Lulus': [1, 0, 1, 0, 1, 0, 1, 0, 1]
})
print(data)
```

Memisahkan Fitur dan Target

```python
X = data[['IPK', 'Kehadiran']]
y = data['Lulus']
```

melatih model

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X, y)
print("Model berhasil dilatih.")
```

Melakukan Prediksi

```python
mahasiswa_baru = pd.DataFrame([[2.0, 75]], columns=['IPK', 'Kehadiran'])
hasil = model.predict(mahasiswa_baru)
print("Prediksi:", hasil)
```
